```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from numpy import random
x=random.uniform(size=(2,3))
print(x)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.uniform(size=1000),hist=False)
plt.show()
```

```python
#logistic array
x=random.logistic(loc=1,scale=3,size=(2,3))
print(x)
```
